```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/Users/aishwaryakompella/Downloads/2301 Week 6 Coding Demo 1-Student/College_Employment_Competitiveness.csv")

# shows first 5 rows 
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df["Employment_Competitiveness"].value_counts()
```

```python
order = ["Low", "Medium", "High"]

df["Employment_Competitiveness"] = pd.Categorical(
    df["Employment_Competitiveness"],
    categories=order,
    ordered=True
)
```

```python
sns.countplot(data=df, x="Employment_Competitiveness", order=order)
plt.title("Distribution of Employment Competitiveness")
plt.xlabel("Competitiveness Level")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(
    data=df,
    x="Employment_Competitiveness",
    y="GPA",
    order=order
)

plt.title("GPA Distribution by Employment Competitiveness")
plt.xlabel("Competitiveness Level")
plt.ylabel("GPA")
plt.show()
```

```python
sns.boxplot(
    data=df,
    x="Employment_Competitiveness",
    y="Exam_Scores",
    order=order
)

plt.title("Exam Scores by Employment Competitiveness")
plt.xlabel("Competitiveness Level")
plt.ylabel("Exam Scores")
plt.show()
```
